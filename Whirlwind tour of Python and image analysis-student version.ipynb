## We can use "conditional" statements to help control what happens in our program

```python
#Example

x = 10.0
y = 20.0

if (x > y):
    print("winner!")
else:
    print("loser!")
    

if (x != y):
    print("really, we're all winners")
```

```python
# We can use a conditional expression in assignments:

# expr1 if condition1 else expr2 will evaluate to "expr1" if condition1 is True; otherwise it evaluates to expr2.

# So for example

x = 2.0
y = 3.0
z = 8.0


x = y if y>z else z
print(x)
x = y if y<z else z
print(x)

```

## Many useful functions and variables are defined by libraries we can "import"
## The two most frequently used in astronomy are numpy and matplotlib



```python
import numpy as np

import matplotlib.pyplot as plt

print(np.pi)

theta = np.linspace(0, 2.0*np.pi, 101) # sets up linearly-spaced values from 0 to 2pi
print(np.min(theta), np.max(theta), np.median(theta))
print(theta[0:10], theta[-10:])
print(theta[-1])
```

```python
# Let's try plotting a sine curve

theta = np.linspace(0, 2.0*np.pi, 101) # sets up linearly-spaced values from 0 to 2pi

y = np.sin(theta)

plt.figure(figsize=(8,8))
plt.plot(theta, y)
plt.title("Our first plot of a sine curve!")
plt.xlabel("angle")
plt.ylabel("function")
```

```python
# There are many, many different equivalent ways to plot data in matplotlib 
# For example, here is a common idiom for doing subplots

fig, axs =plt.subplots(2, 1)

z = np.cos(theta)
axs[0].plot(theta, y)
axs[1].plot(theta, z)

axs[0].set_ylabel("sin")
axs[1].set_ylabel("cos")
axs[0].set_xlabel('theta')
```

```python
#You can actually use plt.subplots() even if you just want one plot, like so:

fig, ax =plt.subplots()

ax.plot(theta, y)
```

## Now let's try some basic image display.  We'll use libraries to help us

```python
from PIL import Image # just one of many possible libraries ("Pillow") for image handling

image_path = 'm51.png'


raw_image = Image.open(image_path).convert('L') # Convert to grayscale ('L' mode)
image_data = np.array(raw_image)

print(f"Image '{image_path}' loaded successfully!")
print(f"Image dimensions: {image_data.shape[1]} pixels (width) x {image_data.shape[0]} pixels (height)")
print(f"Data type: {image_data.dtype}")

plt.figure(figsize=(8, 8))
plt.imshow(image_data, cmap='gray')
plt.title("Grayscale Image of M51 (The Whirlpool Galaxy)")
plt.colorbar(label='Pixel Intensity')


```

## Finally, let's try fitting a simple curve to data using curvefit


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit



# Define the True Function
def linear_function(x, m, c):
    """A simple linear function: y = m*x + c"""
    return m * x + c

# Generate Synthetic Data
# True parameters for our linear function
true_m = 2.5
true_c = 5.0

# Generate x values
x_data = np.linspace(0, 10, 50) # 50 points between 0 and 10

# Generate ideal y values
y_ideal = linear_function(x_data, true_m, true_c)

# Add some random noise to the y values
noise = np.random.normal(0, 1.5, len(x_data)) # Mean 0, std dev 1.5
y_data = y_ideal + noise

# Plot the generated data
plt.figure(figsize=(8, 6))
plt.scatter(x_data, y_data, label='Noisy Data', s=20)
plt.plot(x_data, y_ideal, color='red', linestyle='--', label='Ideal True Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Generated Data with Noise (Linear)')
plt.legend()
plt.grid(True)

# 4. Perform Curve Fitting with curve_fit
# Perform the curve fit
popt, pcov = curve_fit(linear_function, x_data, y_data)

# popt contains the optimal parameters (m_fit, c_fit)
m_fit, c_fit = popt

print("--- Linear Fit Results ---")
print(f"Fitted slope (m): {m_fit:.3f}")
print(f"Fitted intercept (c): {c_fit:.3f}")
print(f"True slope (m): {true_m}")
print(f"True intercept (c): {true_c}")
print("-" * 25)

# 5. Visualize the Fitted Curve
# Generate y values using the fitted parameters
y_fit = linear_function(x_data, m_fit, c_fit)

plt.figure(figsize=(10, 7))
plt.scatter(x_data, y_data, label='Noisy Data', s=20, alpha=0.7)
plt.plot(x_data, y_ideal, color='red', linestyle='--', label='Ideal True Function')
plt.plot(x_data, y_fit, color='green', linewidth=2, label=f'Fitted Function (m={m_fit:.2f}, c={c_fit:.2f})')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear Curve Fitting Demonstration')
plt.legend()
plt.grid(True)
```
